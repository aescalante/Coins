# Labelling
This code creates labels.csv file given a folder with images that is used to train ConvNN for classification

```python
import pandas as pd
import glob

def CreateLabels(InputDir, Type):
    '''This functions creates a InputDir/labels_type.csv file from a given InputDirectory with 
    images of a given type of coin. '''
    
    ValidTypes = ["2euro", "1euro", "50cent", "20cent", "10cent", "5cent", "2cent", "1cent", "Other_NoCoin"]
    
    if Type not in ValidTypes:
        print("Invalidt Type, use a valid type \n")
        print ValidTypes
        return 0
    
    print "Creating labels.csv in folder: ", InputDir," Assumend type: ",  Type
        
    
    CoinorNoCoin = -1
    Group = -1
    CoinType = -1
    
    if Type == "2euro":
        CoinorNoCoin = 0
        Group = 0
        CoinType = 1
        
    elif Type == "1euro":
        CoinorNoCoin = 0
        Group = 0
        CoinType = 0
        
    elif Type == "50cent":
        CoinorNoCoin = 0
        Group = 1
        CoinType = 2
        
    elif Type == "20cent":
        CoinorNoCoin = 0
        Group = 1
        CoinType = 1
        
    elif Type == "10cent":
        CoinorNoCoin = 0
        Group = 1
        CoinType = 0
        
    elif Type == "5cent":
        CoinorNoCoin = 0
        Group = 2
        CoinType = 2
    
    elif Type == "2cent":
        CoinorNoCoin = 0
        Group = 2
        CoinType = 1
        
    elif Type == "1cent":
        CoinorNoCoin = 0
        Group = 2
        CoinType = 0
        
    elif Type == "Other_NoCoin":
        CoinorNoCoin = 1
        Group = -1
        CoinType = -1
    
    images = glob.glob(InputDir+"*.jpg") # Get the coin filenames
    header = "FileName,CoinorNoCoin,Group,CoinType \n"
    labels_file_name = InputDir+"/labels_"+Type+".csv"
    labels_file = open(labels_file_name, 'w')
    labels_file.write(header)
    for k_img in images: 
        line = k_img.split("/")[-1]+","+str(CoinorNoCoin)+","+str(Group)+","+str(CoinType)+"\n"
        labels_file.write(line)
    print(labels_file_name+" Written!! \n")
    labels_file.close()
    
    #Test if the output file can be read as dataframe (useful for debug)
    TestRead = True
    if TestRead == True:
        dfLabels = pd.read_csv(labels_file_name)
        print(dfLabels)
        
#Example    
CreateLabels("ProcessedWhiteBackground/", "1euro")
CreateLabels("ProcessedWhiteBackground/", "2euro")
CreateLabels("ProcessedWhiteBackground/", "50cent")
CreateLabels("ProcessedWhiteBackground/", "20cent")
#CreateLabels("ProcessedWhiteBackground/", "10cent")
CreateLabels("ProcessedWhiteBackground/", "5cent")
#CreateLabels("ProcessedWhiteBackground/", "2cent")
#CreateLabels("ProcessedWhiteBackground/", "1cent")
```
